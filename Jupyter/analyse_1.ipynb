# <p style="font-family: serif; font-size:2.2em; text-align: center;"><u>Normalisation des données</u></p> 

On constate certaines anomalies dans le jeux de données tel que les code postaux, le nom des villes, ext...

```python
# Rectification des code postaux
df["Code postal"] = "0" + df["Code postal"].astype(str)
# On affiche les adresses en minuscule
df["Adresse"] = df["Adresse"].str.lower()
# On affiche le nom des villes en minuscule sauf les premieres lettres
df["Ville"] = df["Ville"].str.title()
# On affiche le nom des entreprises en minuscule sauf les premieres lettres
df["Dénomination"] = df["Dénomination"].str.title()

# Exemple du jeu de données après conversion
df.head()
```
